```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pylab as pl
import matplotlib.gridspec as gridspec
```

```python
df = pd.read_csv('../main/log/incomplete_results.csv', names = ["algo", "experiment_id", "dataset", "seed", "generation", "training_fitness", "timing", "pop_node_count", 
                                                                 "test_fitness", "elite_size", "log_level"])
```

```python
df.head()
```

```python
df.dataset.unique()
```

```python
colors = {"SlimGSGP_1_mul" : "red",
          "SlimGSGP_1_sum":"green",
          "SlimGSGP_2_mul":"blue",
          "SlimGSGP_2_sum":"orange",
          'StandardGP':"black",
          'StandardGSGP':"gray"}
```

```python
for data in df.dataset.unique():
    for algo in df.algo.unique():
        grouped = df.query('algo == @algo and dataset == @data')[['generation', 'pop_node_count', 'elite_size']].groupby('generation').median()
        plt.plot(grouped.index, grouped['elite_size'], label = algo, color = colors[algo])
#     plt.ylim(0,  2000)
    plt.xlim(0,200)
    plt.title(data)
    plt.legend(bbox_to_anchor=(1.1, 1.05))
    plt.show()
```

```python
for data in df.dataset.unique():
    gs = gridspec.GridSpec(2, 2)
    gs.update( wspace=0.4)
    pl.figure()
    for algo in df.algo.unique():

        grouped = df.query('algo == @algo and dataset == @data')[['generation',  "training_fitness","test_fitness", "elite_size"]].groupby('generation').median()
        
        
        ax1 = pl.subplot(gs[0, 0]) 
        pl.plot(grouped.index, grouped['training_fitness'], label = algo, color = colors[algo])
        ax1.set_ylabel('Training Fitness')
        
        
        ax2 = pl.subplot(gs[0, 1]) 
        pl.plot(grouped.index, grouped['test_fitness'], label = algo, color = colors[algo])
        ax2.set_ylabel('Test Fitness')

        ax3 = pl.subplot(gs[1, :]) 
        pl.plot(grouped.index, grouped['elite_size'], label = algo, color = colors[algo])
        ax3.set_ylim(0, 3000)
        ax3.set_ylabel('Elite size')
    
    
    ax1.set_title(data)
    ax3.legend(bbox_to_anchor=(1.1, 1.05))
    plt.show()
```
